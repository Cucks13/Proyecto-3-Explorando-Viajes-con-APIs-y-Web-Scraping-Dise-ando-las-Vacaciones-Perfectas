```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
actividades_berlin = pd.read_csv("../data/ActividadesBerlin.csv")
actividades_viena = pd.read_csv("../data/ActividadesViena.csv")
airbnb_berlin = pd.read_csv("../data/AirBnB_Berlin.csv")
airbnb_viena = pd.read_csv("../data/AirBnB_Viena.csv")
airbnb_berlin = pd.read_csv("../data/AirBnB_Berlin.csv")
vuelos_ida_berlin = pd.read_csv("../data/vuelosIDABERLIN.csv")
vuelos_ida_viena = pd.read_csv("../data/vuelosIDAVIENA.csv")
vuelos_vuelta_berlin = pd.read_csv("../data/vuelosVUELTABERLIN.csv")
vuelos_vuelta_viena = pd.read_csv("../data/vuelosVUELTAVIENA.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(actividades_berlin["Precio Actividad"], bins=10, color="blue", label="Berln", kde=True)
sns.histplot(actividades_viena["Precio Actividad"], bins=10, color="red", label="Viena", kde=True)
plt.title("Distribucin de precios de actividades en Berln vs Viena")
plt.xlabel("Precio Actividad (EUR)")
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.histplot(airbnb_berlin["Precio Total"], bins=10, color="blue", label="Berln", kde=True)
sns.histplot(airbnb_viena["Precio Total"], bins=10, color="red", label="Viena", kde=True)
plt.title("Distribucin de precios de Airbnbs en Berln vs Viena")
plt.xlabel("Precio Total (EUR)")
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(vuelos_ida_berlin["Precio Total (EUR)"], bins=10, color="blue", label="Berln", kde=True)
sns.histplot(vuelos_ida_viena["Precio Total (EUR)"], bins=10, color="red", label="Viena", kde=True)
plt.title("Distribucin de precios de vuelos de ida a Berln vs Viena")
plt.xlabel("Precio Total (EUR)")
plt.legend()
plt.show()
```
